# 使用Baseline算法进行推荐

```python
import pandas as pd
```

```python
movies = pd.read_csv('../database/movies.csv')
movies
```

```python
ratings = pd.read_csv('../database/ratings.csv')
ratings
```

```python
bu = pd.read_csv('bu.csv')
bu
```

```python
bi = pd.read_csv('bi.csv')
bi
```

```python
import numpy as np
```

```python
len(bu)
```

```python
global_mean = ratings['RATING'].mean()
global_mean
```

```python
user_md5 = bu.loc[np.random.randint(0, len(bu)), 'USER_MD5']
user_md5
```

```python
history = ratings[ratings['USER_MD5'] == user_md5]['MOVIE_ID'].values
history
```

```python
user_prediction = []
```

```python
user_bias = bu[bu['USER_MD5'] == user_md5]['BIAS'].values[0]
user_bias
```

```python
for i in range(len(bi)):
    if bi.loc[i, 'MOVIE_ID'] in history:
        continue
    user_prediction.append([bi.loc[i, 'MOVIE_ID'], global_mean + user_bias + bi.loc[i, 'BIAS']])
```

```python
user_prediction
```

```python
user_prediction = pd.DataFrame(user_prediction, columns=['MOVIE_ID', 'RATING'])
user_prediction = user_prediction.sort_values(by='RATING', ascending=False)
user_prediction
```

```python
recommend = user_prediction.iloc[:100]
recommend
```

```python
recommend_movies = recommend['MOVIE_ID'].values
recommend_movies
```

```python
movies.set_index('MOVIE_ID', inplace=True)
movies
```

```python
recommend_movies_info = movies.loc[list(set(recommend_movies.tolist()) & set(movies.index))]
recommend_movies_info
```

```python
recommend_movies_info = recommend_movies_info.reindex(recommend_movies)
recommend_movies_info
```

```python
recommend_movies_info = recommend_movies_info.reset_index()
recommend_movies_info
```

```python
recommend_movies_info = recommend_movies_info[['MOVIE_ID']]
recommend_movies_info
```

```python
baseline_results = {}
```

```python
for i in range(300000):
    user_md5 = bu.loc[i, 'USER_MD5']
    bias = bu.loc[i, 'BIAS']
    recommended_movies = bi[bi['BIAS'] >= (5 - global_mean - bias)]
    history = ratings[ratings['USER_MD5'] == user_md5]['MOVIE_ID'].values
    recommended_movies = recommended_movies.sort_values(by='BIAS', ascending=False)
    tmp = []
    for j, movie_id, bias in recommended_movies.itertuples():
        if movie_id not in history:
            if len(tmp) == 12:
                break
            tmp.append(movie_id)
    baseline_results[user_md5] = tmp
```

```python
baseline_results
```

```python
for i in range(300000, len(bu)):
    user_md5 = bu.loc[i, 'USER_MD5']
    bias = bu.loc[i, 'BIAS']
    recommended_movies = bi[bi['BIAS'] >= (5 - global_mean - bias)]
    history = ratings[ratings['USER_MD5'] == user_md5]['MOVIE_ID'].values
    recommended_movies = recommended_movies.sort_values(by='BIAS', ascending=False)
    tmp = []
    for j, movie_id, bias in recommended_movies.itertuples():
        if movie_id not in history:
            if len(tmp) == 12:
                break
            tmp.append(movie_id)
    baseline_results[user_md5] = tmp
```

```python
baseline_results
```
