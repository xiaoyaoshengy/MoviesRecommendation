# 本文件用于分割数据集，80%的数据用于训练，剩余的构成测试集

```python
import pandas as pd
```

```python
import numpy as np
```

```python
ratings_data = pd.read_csv('../database/ratings.csv')
ratings_data
```

```python
dataset = ratings_data[['USER_MD5', 'MOVIE_ID', 'RATING']]
dataset
```

```python
testset_index = []
```

```python
users_index = dataset.groupby('USER_MD5').any().index
users_index
```

```python
tmp_index = users_index[:10000]
for uid in tmp_index:
    user_rating_data = dataset.where(dataset['USER_MD5'] == uid).dropna()
    index = round(len(user_rating_data) * 0.8)
    testset_index += list(user_rating_data.index.values[index:])
```

```python
testset_index
```

```python
tmp_df = pd.DataFrame(testset_index, columns=['test_index'])
tmp_df
```

```python
tmp_df.to_csv('testset_index_0.csv', index=False)
```
