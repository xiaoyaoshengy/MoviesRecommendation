# 本文件用于整理数据

```python
import pandas as pd
import numpy as np
import json
```

```python
movies_data = pd.read_csv('movies.csv')
movies_data
```

```python
ratings_data = pd.read_csv('ratings.csv')
ratings_data
```

### 整理电影信息数据

```python
movies_data = movies_data[['MOVIE_ID', 'NAME', 'ACTORS', 'DIRECTORS', 'GENRES', 'STORYLINE', 'TAGS']]
movies_data
```

```python
movies_data.set_index('MOVIE_ID', inplace=True)
```

```python
movies_data
```

```python
movies_ratings = ratings_data.groupby('MOVIE_ID').agg(list)[['RATING']]
movies_ratings
```

```python
movies_ratings['AVERAGE_RATING'] = movies_ratings['RATING'].apply(lambda x: sum(x) / len(x))
movies_ratings
```

```python
movies_ratings = movies_ratings[['AVERAGE_RATING']]
movies_ratings
```

```python
movies_process = movies_data.join(movies_ratings)
movies_process
```

```python
with open('images_1.json', 'r') as file:
    movie_imgs = json.loads(file.read())
```

```python
movie_imgs
```

```python
movie_img_data = pd.DataFrame(movie_imgs)
movie_img_data
```

```python
movie_img_data = movie_img_data.T
movie_img_data
```

```python
movie_img_data.reset_index(inplace=True)
movie_img_data
```

```python
movies_process.reset_index(inplace=True)
movies_process
```

```python
movies_process['TAGS'] = movies_process['TAGS'].apply(lambda x: x if type(x) == float else x.split('/'))
movies_process
```

```python
movies_process['GENRES'] = movies_process['GENRES'].apply(lambda x: x if type(x) == float else x.split('/'))
movies_process
```

```python
movies_process = movies_process.join(movie_img_data)
movies_process
```

```python
movies_process.to_json(path_or_buf='movies_processed.json', orient='records')
```
